# PostgreSQL Database Connection & Data Upload

```python
import pandas as pd
from sqlalchemy import create_engine

# Load healthcare dataset
healthcare_df = pd.read_csv("healthcare_dataset.csv")

# Check columns before uploading
print("Dataset Shape:", healthcare_df.shape)
print(healthcare_df.columns)

# PostgreSQL connection
username = "postgres"
password = "your_password_here"   
host = "localhost"
port = "5432"
database = "healthcare_analysis"

engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

# Upload data with original column names
table_name = "patient_records"

healthcare_df.to_sql(
    table_name,
    engine,
    if_exists="replace",
    index=False
)

print(f"Data uploaded successfully into table '{table_name}'")
```

```python
# Verify table
pd.read_sql('SELECT * FROM patient_records LIMIT 5;', engine)
```
