# 1) Import Libraries

```python
# Jupyter notebook display setting
%matplotlib inline

# Core libraries
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

 # 2) Load the Excel File and Read All Required Sheets

```python
# File location
excel_file = "healthcare_dataset.xlsx"

# Read the Excel workbook
workbook = pd.ExcelFile(excel_file)

# Load individual sheets into separate DataFrames
patients_df = pd.read_excel(workbook, sheet_name="Patients_details")
hospitals_df = pd.read_excel(workbook, sheet_name="Hospital_Details")
doctors_df = pd.read_excel(workbook, sheet_name="Doctor_Details")
visits_df = pd.read_excel(workbook, sheet_name="Patients_Data")
```

# 3) Preview the Loaded Data

```python
# Create a dictionary of all DataFrames
data_tables = {
    "Patients Details": patients_df,
    "Hospital Details": hospitals_df,
    "Doctor Details": doctors_df,
    "Patients Data": visits_df
}

# Preview first and last 5 rows of each dataset
for name, df in data_tables.items():
    print(f"\n========== {name} ==========")
    
    print("\nFirst 5 rows:")
    print(df.head())
    
    print("\nLast 5 rows:")
    print(df.tail())
```

# 4) Data Cleaning & Preparation – Merging Datasets

```python
# Step 1: Merge patient visit data with patient basic details
combined_df = visits_df.merge(patients_df, on="P_ID", how="left")

# Step 2: Merge with doctor details
combined_df = combined_df.merge(doctors_df, on="D_ID", how="left")

# Step 3: Ensure column names match before merging hospital data
hospitals_df = hospitals_df.rename(columns={"H_id": "H_ID"})

# Step 4: Merge with hospital details
combined_df = combined_df.merge(hospitals_df, on="H_ID", how="left")

# Preview merged dataset
combined_df.head()
```

```python
# Check structure and data types
combined_df.info()
```

# 5) Descriptive Statistics

```python
# Summary statistics for numeric columns
print(" Numerical Columns Summary:")
display(combined_df.describe())

# Summary statistics for all columns
print("\n Complete Dataset Summary:")
display(combined_df.describe(include="all"))
```

# 6) Data Inspection + Standardization

```python
# DATASET INSPECTION

# Non-null count in each column
print("Non-null values in each column:")
display(combined_df.count())

# Dataset shape
print("\n Dataset shape (rows, columns):")
print(combined_df.shape)

# Data types
print("\n Column data types:")
display(combined_df.dtypes)

# Column names
print("\n Column names:")
display(combined_df.columns)

# STANDARDIZING TEXT COLUMNS

# Columns to clean
text_columns = ["Name", "Doctor", "Hospital"]

# Clean text (remove spaces + proper casing)
for col in text_columns:
    combined_df[col] = combined_df[col].apply(
        lambda x: x.strip().title() if isinstance(x, str) else x
    )

# Preview cleaned data
print("\n Preview after cleaning:")
display(combined_df.head())
```

# 7) Data Integrity Validation + Missing Value Handling

```python
# DATA INTEGRITY VALIDATION FOR KEY COLUMNS

# Check whether key IDs in the merged dataset exist in their master tables
missing_patient_ids = combined_df.loc[~combined_df["P_ID"].isin(patients_df["P_ID"])]
missing_doctor_ids = combined_df.loc[~combined_df["D_ID"].isin(doctors_df["D_ID"])]
missing_hospital_ids = combined_df.loc[~combined_df["H_ID"].isin(hospitals_df["H_ID"])]

print("Invalid Patient IDs:", len(missing_patient_ids))
print("Invalid Doctor IDs:", len(missing_doctor_ids))
print("Invalid Hospital IDs:", len(missing_hospital_ids))
```

```python
# MISSING VALUE SUMMARY

print("Missing Values Summary:")
display(combined_df.isnull().sum())
```

```python
cleaned_df = combined_df.copy()
```

```python
# HANDLE MISSING VALUES

# Create a proper copy before cleaning to avoid SettingWithCopyWarning
cleaned_df = combined_df.copy()

# Drop rows where Patient ID is missing
cleaned_df = cleaned_df.dropna(subset=["P_ID"]).copy()

# Fill missing numeric value in Age with median
cleaned_df["Age"] = cleaned_df["Age"].fillna(cleaned_df["Age"].median())

# Fill missing categorical values with mode
cleaned_df["Test Results"] = cleaned_df["Test Results"].fillna(cleaned_df["Test Results"].mode()[0])
cleaned_df["Name"] = cleaned_df["Name"].fillna(cleaned_df["Name"].mode()[0])
cleaned_df["Gender"] = cleaned_df["Gender"].fillna(cleaned_df["Gender"].mode()[0])
cleaned_df["Blood Type"] = cleaned_df["Blood Type"].fillna(cleaned_df["Blood Type"].mode()[0])

# Check updated missing values
print("Missing Values After Cleaning:")
display(cleaned_df.isnull().sum())
```

# 8) Handling Duplicate Records

```python
# Check duplicate rows
duplicate_count = cleaned_df.duplicated().sum()
print(f"Total duplicate rows found: {duplicate_count}")

# Preview duplicates
duplicates = cleaned_df[cleaned_df.duplicated()]
display(duplicates.head())

# Remove duplicates
cleaned_df = cleaned_df.drop_duplicates()

# Verify
print("Shape after removing duplicates:", cleaned_df.shape)
```

# 9) Converting Data Types

```python
# Convert date columns to datetime format
cleaned_df["Date of Admission"] = pd.to_datetime(
    cleaned_df["Date of Admission"], errors="coerce"
)

cleaned_df["Discharge Date"] = pd.to_datetime(
    cleaned_df["Discharge Date"], errors="coerce"
)

# Convert Billing Amount to numeric
cleaned_df["Billing Amount"] = pd.to_numeric(
    cleaned_df["Billing Amount"], errors="coerce"
)

# Final datatype check
print("Updated Data Types:")
display(cleaned_df.dtypes)
```

# 10) Creating New Derived Columns

```python
# Calculate total length of stay in days
cleaned_df["Length of Stay"] = (
    cleaned_df["Discharge Date"] - cleaned_df["Date of Admission"]
).dt.days

# Create a billing category flag
cleaned_df["High Bill Flag"] = np.where(
    cleaned_df["Billing Amount"] > 30000,
    "High",
    "Normal"
)

# Preview updated dataset
display(cleaned_df.head())
```

# 11) Mapping Categorical Values

```python
# Check unique admission types
print("Unique Admission Types:")
display(cleaned_df["Admission Type"].unique())

# Encode admission types into numeric values
admission_type_map = {
    "Elective": 1,
    "Emergency": 2,
    "Urgent": 3
}

cleaned_df["Admission Type Code"] = cleaned_df["Admission Type"].map(admission_type_map)

# Preview updated dataset
display(cleaned_df.head())
```

# 12) Final Abnormality Checks

```python
# Check for negative length of stay
negative_stay = cleaned_df[cleaned_df["Length of Stay"] < 0]

print("Rows with negative length of stay:")
display(negative_stay)

# Check for future admission dates
future_admissions = cleaned_df[cleaned_df["Date of Admission"] > pd.Timestamp.today()]

print("Rows with future admission dates:")
display(future_admissions)
```
